```python
# Aqui é realizada a importação das bibliotecas
import cv2
import numpy as np
# Caso for usar o Google Colab com a OpenCV, usar a lib abaixo
from google.colab.patches import cv2_imshow
```

```python
# Os 3 códigos abaixo fazem a leitura da imagem para a memória.
# O 0 como segundo argumento da função representa uma FLAG da leitura.
img = cv2.imread('j.png',0)
img_opening = cv2.imread('j_ruido.png',0)
img_closing = cv2.imread('j_furos.png',0)

# Img.shape é uma função que retorna a altura e largura
altura, largura = img.shape

# Numpy Ones, retorna um array com (5 colunas e 5 linhas) preenchido com 1
kernel = np.ones((5,5),np.uint8)
print(kernel)
```

```python
# Ambos os métodos podem ser utilizados em tratamento de imagem para visão computacional
# Erosion - Este método é geralmente utilizado para imagens binárias e serve para remover as bordas da imagem (no caso utilizado aqui).
erosion = cv2.erode(img,kernel,iterations = 2)
# Dilation - Este método funciona de forma inversa ao método anterior, neste caso utilizado para aumentar as bordas da imagem.
dilation = cv2.dilate(img,kernel,iterations = 2)
```

```python
# Ele gera a diferença entre a Erosion e a Dilation, explicado de forma breve.
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
# Este método cria ruidos fora das bordas do J
opening = cv2.morphologyEx(img_opening, cv2.MORPH_OPEN, kernel)
# Este método cria ruidos dentro das bordas do J
closing = cv2.morphologyEx(img_closing, cv2.MORPH_CLOSE, kernel)
```

```python
# Os métodos abaixo servem para exibir a imagem no terminal do Google Colab
cv2_imshow(img)
cv2_imshow(erosion)
cv2_imshow(dilation)
cv2_imshow(opening)
cv2_imshow(closing)
cv2_imshow(gradient)
```
